```python
class Node:
    def __init__(self,data):
        self.left=None
        self.data=data
        self.right=None
    def insertNode(root,data):
        newNode=Node(data)
        if root==None:
            root=newNode
        else:
            if data<root.data:
                root.left= Node.insertNode(root.left,data)
            else:
                root.right= Node.insertNode(root.right,data)
        return root
def preorder(root):
  if root==None:
    return
  print(root.data,end=" ")
  preorder(root.left)
  preorder(root.right)
def postorder(root):
  if root==None:
    return
  preorder(root.left)
  preorder(root.right)
  print(root.data,end=" ")
def inorder(root):
  if root==None:
    return
  preorder(root.left)
  print(root.data,end=" ")
  preorder(root.right)
def bfs(root):
  queue=[]
  queue.append(root)
  while len(queue)!=0:
    curr=queue.pop(0)
    print(curr.data,end=" ")
    if curr.left!=None:
      queue.append(curr.left)
    if curr.right!=None:
      queue.append(curr.right)
  return
def dfs(root):
  stack=[]
  stack.insert(0,root)
  while len(stack)!=0:
    curr=stack.pop(0)
    print(curr.data,end=" ")
    if curr.right!=None:
      stack.insert(0,curr.right)
    if curr.left!=None:
      stack.insert(0,curr.left)
  return
def countnodes(root):
    if root==None:
        return 0
    #count the nodes in left subtree
    n1=countnodes(root.left)
    #count the nodes in right subtree
    n2=countnodes(root.right)
    total=1+n1+n2
    return total
def maxdepth(root):
    if root==None:
        return 0
    lth=maxdepth(root.left)
    rth=maxdepth(root.right)
    totalh=1+max(lth,rth)
    return totalh
def reversetree(root):
    if root==None:
        return
    #step -1 : reverse left subtree
    reversetree(root.left)
    #step -2 :reverse the right subtree
    reversetree(root.right)
    #step -3 :swap root.left with root.right
    root.left,root.right=root.right,root.left
def minnode(root):
    curr=root
    while curr.left!=None:
        curr=curr.left
    return curr
def maxnode(root):
    curr=root
    curr=curr.right
    while curr.right!=None:
        curr=curr.right
    return curr
def searchnode(root,target):
    #if root is none return none
    if root==None or target==root.data:
        return root
    else:
        #search in left subtree
        if target<root.data:
            return searchnode(root.left,target)
        #search in right subtree
        else:
            return searchnode(root.right,target)
#if differnece of heights of left and right subtree is not more than 1 then it is balanced
#a balanced tree is called as AVL tree
#[lh-rh]<=1
def isbalanced(root):
    if root==None:
        return True
    lh=maxdepth(root.left)
    rh=maxdepth(root.right)
    if abs(lh-rh)<=1 and isbalanced(root.left)==True and isbalanced(root.right) ==True:
        return True
    return False
def is_bst(root):
    #if root is none then return true
    if root==None:
        return True
    #if there is left child and data at the left child>data at root then return false
    if root.left!=None and root.left.data>root.data:
        return False
    #if there isa right child and data at this is less than root data then return false
    if root.right!=None and root.right.data<root.data:
        return false
    #if above methods are not false, return true
    if is_bst(root.left)==False and is_bst(root.right)==False:
        return False
    return True
def del_node(root,target):
    parent,curr=None,root
    while (curr!=None and target!=root.data):
        parent=curr
        if target < curr.data:
            curr=curr.left
        else:
            curr=curr.right
    #target is not in the tree
    if curr==None:
        return root
    #node with no kids
    if curr.left==None and curr.right==None:
        if curr==root:
            return None
        if curr==parent.left:
            parent.left=None
        if curr==parent.right:
            parent.right=None
    #node with two kids
    elif curr.left!=None and curr.right!=None:
        replacer=minnode(curr.right)
        temp=replacer.data
        del_node(root,replacer.data)
        curr.data=temp
    else:
        if curr.left!=None:
            kid=curr.left
        else:
            kid=curr.right
        if curr==root:
            root=kid
        if curr==parent.left:
            parent.left=kid
        if curr==parent.right:
            parent.right=kid
    return root
def flatten(root):
    if root==None:
        return None
    if root.left==None and root.right==None:
        return root
    if root.left!=None:
        flatten(root.left)
        temp=root.right #store right subtree address
        root.right=root.left #connect lst in place of rst
        root.left=None #remove lst
        #go to right most node and connect previous right subtree there
        curr=root
        while curr.right!=None:
            curr=curr.right
        #connect removes rst at the right most end of new rst
        curr.right=temp
    flatten(root.right)


root=None
root=Node.insertNode(root,100)
Node.insertNode(root,95)
Node.insertNode(root,90)
Node.insertNode(root,85)
Node.insertNode(root,110)
Node.insertNode(root,105)
Node.insertNode(root,115)
#Node.insertNode(root,75)
#Node.insertNode(root,130)
#Node.insertNode(root,125)
print("\ninorder")
inorder(root)
print("\npre order")
preorder(root)
print("\npost order")
postorder(root)
print("\nbredth first")
bfs(root)
print("\ndepth first")
dfs(root)
print("\nno of nodes:")
countnodes(root)
print("\nmaximum depth of the tree:",maxdepth(root))
print("\nbefore reversing")
inorder(root)
reversetree(root)
print("\nafter reversing")
inorder(root)
print("\n maximum node : ",maxnode(root).data)
print("\n minnode node : ",minnode(root).data)
node=searchnode(root,120)
print(node)
if isbalanced(root)==True:
    print("It is a balanced tree")
else:
    print("not a balanced tree")
print(is_bst(root))
inorder(root)
flatten(root)
print("\nafter flatttening\n")
inorder(root)
```
